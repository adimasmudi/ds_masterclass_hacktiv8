````python
import pandas as pd

df_jumlah_guru = pd.read_csv("data/jumlah_guru_SD_jakarta.csv")


df_jumlah_guru.head()

````

````python
df_jumlah_guru.negeri = df_jumlah_guru.negeri.str.replace(',','').astype(int)
````

````python
df_jumlah_sekolah =  pd.read_csv("data/jumlah_sekolah_SD_jakarta.csv")
df_jumlah_sekolah.info()
````

````python
df_jumlah_guru.sum()
````

````python
df_kotif_guru = df_jumlah_guru[['kabkota','negeri','swasta']]
````

````python
df_jumlah_sekolah.columns = ['kabkota', 'kecamatan', 'sekolah negeri', 'sekolah swasta']
````

````python
df_jumlah_sekolah['guru negeri'] = df_jumlah_guru['negeri']
df_jumlah_sekolah['guru swasta'] = df_jumlah_guru['swasta']
````

````python
df_jumlah_sekolah.head()
````

````python
df_jumlah_sekolah['rerata_guru_negeri_per_sekolah'] = df_jumlah_sekolah['guru_negeri'] / df_jumlah_sekolah['sekolah negeri']
df_jumlah_sekolah['rerata_guru_swasta_per_sekolah'] = df_jumlah_sekolah['guru_swasta'] / df_jumlah_sekolah['sekolah swasta']
````

````python
df_jumlah_sekolah.head()
````

````python
df_jumlah_sekolah.sort_values('rerata_guru_negeri_per_sekolah',ascending=False,inplace=True)
````
